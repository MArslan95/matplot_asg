```python
import matplotlib.pyplot as plt
import numpy as np

# (1) Create one random graph with plt.plot
x = np.linspace(0, 10, 100)
y = np.random.rand(100)
plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.title('Random Plot')
plt.show()
```

```python

# (2) Create one random graph with plt.bar
x = ['A', 'B', 'C', 'D', 'E']
y = np.random.rand(5)
plt.figure(figsize=(8, 6))
plt.bar(x, y)
plt.title('Random Bar Graph')
plt.show()
```

```python
# (3) Create one random graph with plt.scatter
x = np.random.rand(50)
y = np.random.rand(50)
plt.figure(figsize=(8, 6))
plt.scatter(x, y)
plt.title('Random Scatter Plot')
plt.show()
```

```python
# (4) Create one random graph with plt.hist
data = np.random.normal(0, 1, 1000)
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, alpha=0.5)
plt.title('Random Histogram')
plt.show()
```

```python
# (5) Use plt.subplots to insert two of the above graphs into one figure.
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot 1 - Random Bar Graph
x = ['A', 'B', 'C', 'D', 'E']
y = np.random.rand(5)
axes[0].bar(x, y)
axes[0].set_title('Random Bar Graph')

# Plot 2 - Random Scatter Plot
x = np.random.rand(50)
y = np.random.rand(50)
axes[1].scatter(x, y)
axes[1].set_title('Random Scatter Plot')

plt.show()
```

```python
# (6) Create one random image with plt.imshow()
random_image = np.random.random((30, 30))
plt.figure(figsize=(8, 6))
plt.imshow(random_image, cmap='viridis')
plt.title('Random Image')
plt.show()
```

```python
# (7) Arbitrary function and its derivative
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
dy = np.cos(x)

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, y, label='sin(x)')
ax.plot(x, dy, label="cos(x)")
plt.legend()
plt.title('Plot of sin(x) and its derivative cos(x)')
plt.show()
```
