```python
import matplotlib.pyplot as plt
import numpy.random as rnd
```

## 1. Creating a random random graph with plt.plot 

```python
x =  rnd.randn(100)
y = rnd.randn(100)
# using plt function to create a graph
plt.figure(figsize=(8,6))
plt.plot(x,y,'o')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Random Graph using Plt.plot')
plt.show()
```

## 2. Create One random graph with plt.bar

```python
x = rnd.randint(10, size=10)
y = rnd.randint(10, size=10)
plt.bar(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Random Graph using Plt.bar')
plt.show()
```

## 3. Creating random graph with plt.Scatter

```python
x =  rnd.randn(100)
y = rnd.randn(100)
# using plt to create a graph
plt.figure(figsize=(8,6))
plt.scatter(x,y,c='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Random Graph using Plt.Scater')
plt.show()
```

## 4. Creating random graph using Hist

```python
hist_x = rnd.randn(1000)

plt.figure(figsize = (8,6))
plt.hist(hist_x)
plt.xlabel('Hist_X')
plt.ylabel('Frequency')
plt.title('Random Graph with plt.hist')
plt.show()
```

##  (5) Use plt.subplots to insert two of the above graphs into one figure

```python
fig, axes = plt.subplots(2, 1, figsize=(8, 6))

# Plot a random scatter plot on the first axis
axes[0].scatter(rnd.randn(100), rnd.randn(100), c='red')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].set_title('Random Scatter Plot')

# Plot a random bar plot on the second axis
axes[1].bar(rnd.randint(10, size=10), rnd.randint(10, size=10))
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')
axes[1].set_title('Random Bar Plot')

fig.suptitle('Two Random Graphs in One Figure')
plt.show()
```

## (6) Create one random image with plt.imshow()


```python
import numpy as np
# Create a random image with NumPy
image = rnd.randn(100, 100)

# Plot the image with Matplotlib
plt.figure(figsize=(8, 6))
plt.imshow(image, cmap='gray')
plt.colorbar()
plt.title('Random Image with plt.imshow()')
plt.show()
```

## (7) Determine an arbitrary function and create a function that differentiates it. Display these two together on one graph and write down what the differential value means.


```python

# Define an arbitrary function
def f(x):
  return x**2 + 2*x + 1

# Define a function that differentiates the given function
def f_prime(x):
  return 2*x + 2

# Generate a range of x-values
x = np.linspace(-5, 5, 100)

# Calculate the function values and its derivative
y = f(x)
y_prime = f_prime(x)

# Plot the function and its derivative on the same graph
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x)')
plt.plot(x, y_prime, label='f\'(x)')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Function and Derivative')
plt.legend()
plt.show()
```
